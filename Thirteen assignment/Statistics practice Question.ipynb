````python
# 3.Write a Python class representing a discrete random variable with methods to calculate its expected value and variance.

values=[1,2,3,4,5,]
probabilities=[0.1, 0.2, 0.3, 0.2, 0.2]

class DiscreteRandomVariable:
    def __init__(self,values,probabilities):
        self.values=np.array(values)
        self.probabilities=np.array(probabilities)
        
    def expected_value(self):
        return np.sum(self.values*self.probabilities)
    
    def variance(self):
        mean= self.expected_value()
        return np.sum(self.probabilities*(self.values-mean)**2)
    
    
random_variable = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", random_variable.expected_value())
print("Variance:", random_variable.variance())
````

````python
# 5) Create a Python function to generate random samples from a given probability distribution (e.g., 
# binomial, Poisson) and calculate their mean and variance.

def generate_random_sample(distribution,param,sample_size=1000):
    if distribution == 'binomial':
        n=param.get("n",1)
        p=param.get("p",0.5)
        sample=np.random.binomial(n,p,sample_size)
    elif distribution == 'poisson':
        lam=param.get("lam",1)
        sample=np.random.poisson(lam,sample_size)
    else:
        raise ValueError("Invalid distribution")
    mean=np.mean(sample)
    variance=np.var(sample)
    return(mean,variance)

binomial_params={"n":10,"p":0.5}
poisson_params={"lam":5}
            
binomial_mean ,binomial_variance=generate_random_sample("binomial",binomial_params)
poisson_mean ,poisson_variance=generate_random_sample("poisson",poisson_params)
print("Binomial Distribution - Mean:", binomial_mean, "Variance:", binomial_variance)
print("Poisson Distribution - Mean:", poisson_mean, "Variance:", poisson_variance)
````

````python
# 11. A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors 
# who make a purchase). They collect data from the old and new layouts to compare.


import numpy as np 
# 50 purchases out of 1000 visitors 
old_layout = np.array([1] * 50 + [0] * 950) 
# 70 purchases out of 1000 visitors   
new_layout = np.array([1] * 70 + [0] * 930) 

# Import necessary libraries
from statsmodels.stats.proportion import proportions_ztest

# Calculate the number of successes and total trials
successes = np.array([np.sum(old_layout), np.sum(new_layout)])
trials = np.array([len(old_layout), len(new_layout)])

# Perform the Z-test for proportions
stat, p_value = proportions_ztest(successes, trials, alternative='larger')

# Display the results
print(f"Z-statistic: {stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of results
if p_value < 0.05:
    print("Reject the null hypothesis: The new layout has a significantly higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

````python
# 15. A company is testing two different website layouts to see which one leads to higher click-through rates. 
# Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of 
# freedom, and p-value.
"""  

Use the following data: 
```python 
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37] 
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

"""

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37] 
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

from scipy.stats import ttest_ind

def ab_test(layout_a_clicks, layout_b_clicks):
    layout_a_clicks = np.array(layout_a_clicks)
    layout_b_clicks = np.array(layout_b_clicks)
    t_stat, p_value=ttest_ind(layout_a_clicks,layout_b_clicks)
    
    df =len(layout_a_clicks) + len(layout_b_clicks) - 2
    
    print(f"t-statistic: {t_stat:.4f}")
    print(f"Degrees of freedom: {df}")
    print(f"P-value: {p_value:.4f}")
    
    if p_value < 0.05:
        print("Reject the null hypothesis. There is a statistically significant difference between the two layouts.")
        
    else:
        print("Fail to reject the null hypothesis. There is no statistically significant difference between the two layouts.")
        
ab_test(layout_a_clicks, layout_b_clicks)
````

````python
# 21. A political analyst wants to determine if there is a significant association between age groups and voter 
# preferences  (Candidate A or Candidate B). They collect data from a sample of 5 00 voters and classify 
# them into different age groups and candidate preferences. Perform a Chi-S quare test to determine if 
# there is a significant association between age groups and voter preferences.  

from scipy.stats import chi2_contingency

np.random.seed(0) 
age_groups = np.random.choice(['18 -30', '31 -50', '51 +', '51 +'], size=30) 
voter_preferences = np.random.choice(['C andidate A', 'C andidate B'] , size=30) 

data=pd.DataFrame({'Age Group':age_groups,'Voter Preferences':voter_preferences})
contingency_table=pd.crosstab(data['Age Group'],data['Voter Preferences'])

chi2,p_value,dof,expected=chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n",expected)

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis. No significant association between age groups and voter preferences.")
````

````python
from scipy.stats import f_oneway

# 24. A company produces three different versions of a product:  Standard, Premium, and Deluxe. The 
# company wants to determine if there is a significant difference in customer satisfaction scores among the 
# three product versions. They conducted a survey and collected customer satisfaction scores for each 
# version from a random sample of customers. Perform an ANOV A test to determine if there is a significant 
# difference in customer satisfaction scores.  

  # Sample data: Customer satisfaction scores for each product version 
  
from scipy.stats import f_oneway

standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87] 
premium_scores = [90, 92, 88, 92, 95, 91 , 96, 93, 89, 93] 
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99] 

f_stat, p_value = f_oneway(standard_scores, premium_scores, deluxe_scores)

print(f"F-statistic: {f_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject the null hypothesis. There is a significant difference in customer satisfaction scores among the product versions.")
else:
    print("Fail to reject the null hypothesis. No significant difference in customer satisfaction scores among the product versions.")
````
